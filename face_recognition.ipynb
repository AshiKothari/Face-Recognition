```python
import cv2
```

```python
img=cv2.imread('ashi.jpeg')
nimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
```

```python
# reading by matplotlib(reads img as rgb) so we converted bgr to rbg (opencv reads rgb images as bgr images)
plt.imshow(img)
plt.show()
plt.imshow(nimg)
plt.show()
```

```python
# reading by opencv it reads bgr as bgr so no need for conversion
cv2.imshow('Ashi',img)
cv2.waitKey(0) 
#time in ms
cv2.destroyAllWindows()
```
